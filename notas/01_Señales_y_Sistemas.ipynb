```python
import numpy as np
import sympy as sym
from sympy.plotting import plot

%matplotlib inline
import matplotlib.pyplot as plt
```

# Introducción a Señales y Sistemas

```python
def delta_n(n):
    return 0. if n != 0 else 1.

n_range = np.arange(-4,4)
a = -2 * np.array([delta_n(n+2) for n in n_range])
fig, ax = plt.subplots(1,1,figsize=(5,2))
ax.stem(n_range,a,use_line_collection=True)
ax.grid(True)
```

```python
a_int = np.cumsum(a)
fig, ax = plt.subplots(1,1,figsize=(5,2))
ax.stem(n_range,a_int,use_line_collection=True)
ax.grid(True)
```

```python
def step(n):
    return 1. if n >= 0 else 0

n_range = np.arange(-3,4)
b = np.array([2.**n * step(n) for n in n_range])
fig, ax = plt.subplots(1,1,figsize=(5,2))
ax.stem(n_range,b,use_line_collection=True)
ax.grid(True)
```

```python
def delta(n_range):
    x = np.zeros(n_range.shape)
    x[n_range == 0] = 1.
    return x

def step(n_range):
    x = np.zeros(n_range.shape)
    x[n_range >= 0] = 1.
    return x

n_range = np.arange(-10,20)

fig, ax = plt.subplots(5,1,figsize=(7,6))
ax[0].stem(n_range,2*delta(n_range-4),use_line_collection=True)
ax[0].grid(True)

ax[1].stem(n_range,step(n_range+2),use_line_collection=True)
ax[1].grid(True)

ax[2].stem(n_range,2.**n_range * step(n_range+2),use_line_collection=True)
ax[2].grid(True)

ax[3].stem(n_range,2.**(-n_range) * step(-n_range),use_line_collection=True)
ax[3].grid(True)

ax[4].stem(n_range,np.cos(np.pi/3 *n_range)*step(n_range-2),use_line_collection=True)
ax[4].grid(True)



fig.tight_layout()
```

```python


def x(t):
    if t <= 0:
        return 0
    elif t > 0 and t <=10:
        return 1
    elif t > 10 and t <= 20:
        return 2 - 0.1 * t
    

t = sym.symbols("t")
x = lambda t: sym.Piecewise( (0, t <= 0), (1, sym.And(t > 0, t <=10)), (2 - t / 10, sym.And(t > 10, t <= 20)) )
# print(x.subs(t,-1), x.subs(t,5), x.subs(t,15))
p = plot(x(t),(t,-5,50),xlim=(-5,50),show=False)
p.append(
    plot(x(t-3),(t,-5,50),xlim=(-5,50),line_color="tab:orange",show=False)[0]
)
p.show()
```
